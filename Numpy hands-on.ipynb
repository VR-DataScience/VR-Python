### Numpy Broadcasting
Broadcasting is used to describe how arrays are treated during mathematical operations.
The term broadcast is used because a small array is stretched or *broadcasted* over a larger array so that they have compatible sizes.

** Broadcasting Rules: **
Starting from the last axis and working backwards, Numpy compares the array dimensions:
* If two dimensions are __equal__, you can __continue__     (Size of each dimension should be same)
* If one of the __operand is 1__, __stretch__ it to match the largest one     (Size of one of the dimension should be 1)
* When one of the __shapes run out of dimension__ (because it has less dimensions than the other shape), Numpy will __assign 1__ in the comparision process until the other shape's dimensions run out as well. 


```python
arr_2d = np.array([[1,2,3],[5,6,7],[8,9,10],[12,13,14]])    # Creating numpy array
arr_2d
```

```python
arr_2d.shape
```

```python
scaler = 3                                                  # scaler
```

```python
scaler
```

```python
arr_2d +  3                                            # operation with a scaler
```

arr_2d --> 4 * 3

scaler --> 1 * 1<br>
Here the scaler is stretched and is added to all elements of the array. 

```python
arr_1d = np.array([10,10,10])                               # array with different shape
arr_1d.shape
```

```python
print(arr_1d)
print()
print()
print(arr_2d)
```

```python
arr_2d + arr_1d                           # operation with a array of different shape
```

arr_2d --> 4 * 3

arr_1d --> 1 * 3
In this case the array with lower dimension i.e. arr_1d is stretched such that it matches the dimensions of arr_2d
And then addition is performed. Lets learn it better with one diagramatic example.

## Power of Matrix 

```python
# np.linalg.matrix_power(a,n)     - a is matrix and n is power
# Matrix need to be square matrix
# n can be positive, negative and zero
# if n = 0, then Identity matrix will be result
#if n > 0, Power is computed through repeated matrix squaring and matrix multiplication
# if n < 0, first we will find out inverse of a matrix, then take absoulte value of inverse matrix and then raise the inverse matrix to power n
```

```python
a = np.arange(1,5).reshape(2,2)
a
```

```python
#if n> 0
np.linalg.matrix_power(a,2)
```

```python
# check above answer
np.dot(a,a)
```

```python
np.linalg.matrix_power(a,3)
```

```python
# if n= 0    - result will be identity matrix
np.linalg.matrix_power(a,0)
```

```python
# if n < 0 
np.linalg.matrix_power(a,-2)
```

```python
# check inverse of matrix a 
b = np.linalg.inv(a)
b
```

```python
np.linalg.matrix_power(b,2)   # this is same as above answer
```

```python
np.dot(b,b)     # double check through dot function
```

## Linear Equation

```python
# 3x + y = 9
# x + 2y = 8
# Find the value of x and y basis given linear equations
```

```python
# A x = B
# A is Matrix of coefficient
# x is matrix of unknown variable
# B is the constant values 
```

```python
# A in above given linear equations
a = np.array([[3,1],[1,2]])
b = np.array([9,8])

np.linalg.solve(a,b)      # return result will be identical to shape of b
```

```python
# Result ----->   x =2 and y = 3
```

```python
# 3 equations example
# 6x + 2y - 5z = 13
# 3x + 3y - 2z = 13
# 7x + 5y - 3z = 26

a = np.array([[6,2,-5],[3,3,-2],[7,5,-3]])
b = np.array([13,13,26])

np.linalg.solve(a,b)   # Result [ x =2, y=3, and z= 1 ]
```

```python
c = np.array([[2,1,3],[1,-1,-1],[3,-2,3]])
d = np.array([-2,-3,-12])

np.linalg.solve(c,d)  
```
